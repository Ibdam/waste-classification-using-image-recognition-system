```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

test_datagen = ImageDataGenerator(rescale=1./255)

test_data = test_datagen.flow_from_directory(
    'waste_dataset_split/test',
    target_size=(150,150),
    batch_size=32,
    class_mode='categorical',
    shuffle=False   # ⚠️ Important: don’t shuffle test data
)
```

```python
from tensorflow.keras.models import load_model

model = load_model('best_waste_model.h5')
```

```python
test_loss, test_accuracy = model.evaluate(test_data)
print("Test Accuracy:", test_accuracy)
print("Test Loss:", test_loss)
```

```python
import numpy as np

predictions = model.predict(test_data)
predicted_classes = np.argmax(predictions, axis=1)

# Get true labels and class names
true_classes = test_data.classes
class_labels = list(test_data.class_indices.keys())

print("Predicted class names:", [class_labels[i] for i in predicted_classes[:10]])
```
